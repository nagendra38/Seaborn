```python
import seaborn as sns

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
penguins.shape
```

```python
sns.set_style('darkgrid')
```

```python
sns.barplot(x='species', y='body_mass_g', data=penguins);
```

```python
penguins.groupby('species').body_mass_g.mean()
```

```python
sns.barplot(x='body_mass_g', y='species', data=penguins);
```

```python
 sns.barplot(x='species', y='body_mass_g', 
            data=penguins,
            ci=68
           );
```

```python
sns.barplot(x='species', y='body_mass_g', 
            data=penguins,
            ci=None
           );
```

```python
sns.barplot(x='species', y='body_mass_g', 
            data=penguins,
            estimator=np.std
           );
```

```python
 penguins.groupby('species').body_mass_g.describe()
```

```python
sns.barplot(x='species', y='body_mass_g', 
            data=penguins,
            estimator=np.max
           );
```

```python
 sns.barplot(x='species', y='body_mass_g', 
            data=penguins,
            hue='sex'
           );
```

```python
sns.barplot(x='body_mass_g', y='species',  
            data=penguins,
            hue='sex'
           );
```

```python
 sns.barplot(x='species', y= 'body_mass_g', 
            data=penguins,
            order=['Gentoo', 'Chinstrap', 'Adelie']
           );
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()
```

```python
cars.shape
```

```python
cars.head()
```

```python
sns.set_style('whitegrid')
```

```python
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
sns.violinplot(x='displacement',data=cars)
```

```python
 sns.violinplot(x=cars.cylinders,y=cars.displacement)
```

```python
  sns.violinplot(x='cylinders',y='displacement',data=cars) 
```

```python
  sns.violinplot(x='cylinders',y='displacement',hue='origin',data=cars) 
```

```python
   sns.violinplot(x='cylinders',y='displacement',hue='origin',data=cars[cars.origin.isin(['japan','europe'])]) 
```

```python
    sns.violinplot(x='cylinders',y='displacement',hue='origin',data=cars[cars.origin.isin(['japan','europe'])],split=True) 
```

```python
    sns.violinplot(x='cylinders',y='displacement',hue='origin',data=cars[cars.origin.isin(['japan','europe'])],split=True,inner='quartiles') 
```

```python
sns.violinplot(x='cylinders',y='displacement',hue='origin',
               data=cars[cars.origin.isin(['japan','europe'])],
               split=True,
              scale='count') 
```

```python
sns.violinplot(x='cylinders',y='displacement',hue='origin',
               data=cars[cars.origin.isin(['japan','europe'])],
               split=True,
              scale='count',
              scale_hue=False) 
```

```python
cars[cars.origin.isin(['japan','europe'])].groupby('cylinders').origin.value_counts()
```
